```python
import requests
from bs4 import BeautifulSoup

url = input("Enter the URL of the website: ")

try:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    links = soup.find_all('a')
    
    for link in links:
        href = link.get('href')
        if href:
            print(href)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from bs4 import BeautifulSoup

html_content = """
<html>
    <head>
        <title>Page title</title>
    </head>
    <body>
        <p id="firstpara" align="center">This is paragraph<b>one</b></p>
        <p id="secondpara" align="blah">This is paragraph<b>two</b></p>
    </body>
</html>
"""

# i) Read the page using BeautifulSoup and show it in a well-formatted, indented manner
soup = BeautifulSoup(html_content, 'html.parser')
print(soup.prettify())

# ii) Print the 'b' tag from the page
b_tag = soup.find('b')
if b_tag:
    print("ii) Found 'b' tag:", b_tag.text)

# iii) Print all the tags that start from 'b'
tags_starting_with_b = soup.find_all(lambda tag: tag.name.startswith('b'))
if tags_starting_with_b:
    print("iii) Tags starting with 'b':", [tag.name for tag in tags_starting_with_b])

# iv) Print text from the tags having 'title' and 'p' by using lists
title_and_p_tags = soup.find_all(['title', 'p'])
if title_and_p_tags:
    print("iv) Text from 'title' and 'p' tags using lists:")
    for tag in title_and_p_tags:
        print(tag.text)

# v) Print text from the tags having 'title' and 'p' by using dictionaries
title_and_p_tags = soup.find_all({'title': True, 'p': True})
if title_and_p_tags:
    print("v) Text from 'title' and 'p' tags using dictionaries:")
    for tag in title_and_p_tags:
        print(tag.text)

# vi) Print all the tag names present in the page
tag_names = [tag.name for tag in soup.find_all()]
if tag_names:
    print("vi) All tag names:", tag_names)

# vii) Print the complete tag that has two, and only two, attributes
tags_with_two_attributes = soup.find_all(lambda tag: len(tag.attrs) == 2)
if tags_with_two_attributes:
    print("vii) Tags with two attributes:")
    for tag in tags_with_two_attributes:
        print(tag)

# viii) Print the tags that have one-character names and no attributes
tags_with_one_character_names = soup.find_all(lambda tag: len(tag.name) == 1 and not tag.attrs)
if tags_with_one_character_names:
    print("viii) Tags with one-character names and no attributes:", [tag.name for tag in tags_with_one_character_names])

# ix) Print all the tags which have a value of 'center' for the 'align' attribute
tags_with_center_align = soup.find_all(align='center')
if tags_with_center_align:
    print("ix) Tags with 'align' attribute equal to 'center':")
    for tag in tags_with_center_align:
        print(tag)

# x) From the XML content '<person name="Bob"><parent rel="mother" name="Alice">'
#    print the attributes having the name as 'Alice'
import xml.etree.ElementTree as ET
xml_content = '<person name="Bob"><parent rel="mother" name="Alice"></parent></person>'
root = ET.fromstring(xml_content)
for elem in root.iter():
    if 'name' in elem.attrib and elem.attrib['name'] == 'Alice':
        print("x) Attributes with name='Alice':", elem.attrib)
```
