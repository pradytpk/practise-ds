# Matplot 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 5.1: Draw a simple plot
X = [1, 2, 3, 4]
y = [20, 21, 20.5, 20.8]
plt.figure()
plt.plot(X, y)
plt.title('Simple Plot')
plt.show()
```

```python
# 5.2: Configure the line and markers in a simple plot
plt.figure()
plt.plot(X, y, linestyle='--', marker='o', color='r')
plt.title('Plot with Line and Markers')
plt.show()
```

```python
# 5.3: Configure the axes
plt.figure()
plt.plot(X, y)
plt.xlim(0, 5)
plt.ylim(19, 22)
plt.title('Plot with Configured Axes')
plt.show()
```

```python
# 5.4: Give the title to the graph & labels of the X-axis and Y-axis
plt.figure()
plt.plot(X, y)
plt.title('Graph Title')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.show()
```

```python
# 5.5: Give the error bar if y_error = [0.12, 0.13, 0.2, 0.1]
y_error = [0.12, 0.13, 0.2, 0.1]
plt.figure()
plt.errorbar(X, y, yerr=y_error, fmt='-o')
plt.title('Plot with Error Bars')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 5.6: Define width, height as figsize=(4,5) DPI and adjust plot dpi=100
plt.figure(figsize=(4, 5), dpi=100)
plt.plot(X, y)
plt.title('Plot with Custom Figure Size and DPI')
plt.show()
```

```python
# 5.7: Give a font size of 14
plt.figure()
plt.plot(X, y)
plt.title('Plot with Custom Font Size', fontsize=14)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.show()
```

```python
# 5.8: Draw a scatter graph of any 50 random values across X and Y axes
X_rand = np.random.rand(50)
y_rand = np.random.rand(50)
plt.figure()
plt.scatter(X_rand, y_rand)
plt.title('Scatter Plot of 50 Random Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python

# 5.9: Create a DataFrame
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}
df = pd.DataFrame(data)

plt.figure()
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'], alpha=0.5)
plt.title('Scatter Plot of preTestScore vs postTestScore (Size by Age)')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
# 5.10: Draw a Scatterplot from the data in question 9
plt.figure()
colors = df['female'].map({0: 'blue', 1: 'red'})
plt.scatter(df['preTestScore'], df['postTestScore'], s=300, c=colors)
plt.title('Scatter Plot of preTestScore vs postTestScore (Color by Sex)')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
```

```python
# 1.You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the 
# United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a 
# Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:\Data Science\Datasets\Hurricanes.csv")

x = df["Year"]
y = df["Hurricanes"]

plt.bar(x,y)
plt.xlabel("Year")
plt.ylabel("Hurricanes")
plt.grid()
plt.title("Hurricanes vs Year")

plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right') # Rotate Axis Labels

plt.show()
```

```python
# 2.The dataset given, records data of city temperatures over the years2014 and 2015. 
# Plot the histogram of the temperatures over this period for the cities of San Francisco and Moscow.

import pandas as pd
import matplotlib.pyplot as plt

df_temp = pd.read_csv("D:\Data Science\Datasets\CityTemps.csv")
print(df_temp)

x_mosco = df_temp["Moscow"]
x_san_Francisco = df_temp["San Francisco"]

plt.hist(x_mosco, len(x_mosco), facecolor='blue', alpha=0.5)
plt.hist(x_san_Francisco, len(x_san_Francisco), facecolor='red', alpha=0.5)

plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Temperatures of Mosco and San Francisco")

plt.setp(plt.gca().get_xticklabels(), rotation=90, horizontalalignment='right') # Rotate Axis Labels

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:\Data Science\Datasets\BigMartSalesData.csv")

df_grouped_year = df.query("Year == 2011").filter(
    ["Month", "Amount"]).groupby(["Month"], as_index=False).sum()

# Plot Total Sales Per Month for Year 2011.
x = df_grouped_year["Month"]
y = df_grouped_year["Amount"]
plt.plot()
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Month vs sales")
plt.grid()
plt.show()

# How the total sales have increased over months in Year 2011.
# Total sales almost increased till 11th month but decreased in 12th month

# Which month has the lowest Sales?
# By seeing the graph, we can directly say that the lowest sales are in the 2nd month
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("D:\Data Science\Datasets\BigMartSalesData.csv")

df_grouped_year = df.query("Year == 2011").filter(
    ["Month", "Amount"]).groupby(["Month"], as_index=False).sum()

# Plot Total Sales Per Month for Year 2011 as Bar Chart.
x = df_grouped_year["Month"]
y = df_grouped_year["Amount"]

plt.plot(x, y)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Month vs sales")
plt.grid()
plt.show()

# Is Bar Chart Better to visualize than Simple Plot?

# Line graphs are used to track changes over short and long periods of time.
# When smaller changes exist, line graphs are better to use than bar graphs.
# Line graphs can also be used to compare changes over the same period of time for more than one group.

# Bar graphs are used to compare things between different groups or to track changes over time.
# However, when trying to measure change over time, bar graphs are best when the changes are larger.
```

```python
# Plot Pie Chart for Year 2011 Country Wise.

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:\Data Science\Datasets\BigMartSalesData.csv")

df_grouped_country = df.query("Year == 2011").filter(
    ["Country", "Amount"]).groupby(["Country"], as_index=False).sum()

plt.pie(df_grouped_country["Amount"],
        labels=df_grouped_country["Country"],
        autopct='%1.1f%%',
        shadow=True)

plt.tight_layout()
plt.show()

# Which Country contributes highest towards sales?
# United Kingdom
```

```python
# 4. Plot Scatter Plot for the invoice amounts and see the concentration of amount.
import pandas as pd

import matplotlib.pyplot as plt
df = pd.read_csv("D:\Data Science\Datasets\BigMartSalesData.csv")

invoice_amounts = df.filter(["InvoiceDate", "Amount"]).groupby(
    "InvoiceDate", as_index=False).sum()

plt.scatter(invoice_amounts["InvoiceDate"], invoice_amounts["Amount"])
plt.ylim(20000, 100000)
plt.show()

# In which range most of the invoice amounts are concentrated
# 20000 - 30000
```
